```python
import matplotlib.pyplot as plt 
import numpy as np 
```

plotting

```python
x=np.arange(0,10)
y=x*2
print("our x is{0} and our y is {1}".format(x,y))
```

```python
plt.plot(x,y)
plt.title("first figure")
plt.xlabel("x axis")
plt.ylabel('y=x^2')
plt.xlim(0,9)             # deterimine the range of x axis     x limitation 
plt.ylim(0,12) 
plt.savefig(r'C:\Users\Amir\Desktop\myploy.png')
```

```python
a=np.linspace(0,12,10)
b=a**3
print('a is \n {0} \n and b is \n {1}'.format(a,b))
```

```python
fig=plt.figure()
axes1=fig.add_axes([0,0,1,1])
axes1.plot(a,b)

axes2=fig.add_axes([0.2,0.2,0.3,0.3])
axes2.plot(x,y)
```

SUBPLOTTING


```python
fig=plt.figure(figsize=(5,5), dpi=200)
axes1=fig.add_axes([0,0,1,1])
axes1.plot(a,b)
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[1].plot(a,b)
```

```python
fig,axes=plt.subplots(nrows=2,ncols=2)
axes[0][1].plot(x,y)
axes[1][0].set_title('x**3')
axes[1][0].plot(a,b)
```

legend

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,label='y=x*2')
ax.plot(a,b,label='y=x**3')
ax.legend()
```
