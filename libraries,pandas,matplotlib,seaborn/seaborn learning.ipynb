```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
df=pd.read_csv(r'HousePrice.csv')
df.head()
```

```python
sns.scatterplot(x='area', y='Price', data=df, hue='Room', palette='Dark2', size='Room', style='Parking')
```

```python
sns.rugplot(x='area', data=df, height=0.7)
```

```python
sns.displot(x='area', data=df)
```

```python
sns.displot(x='area', data=df, kde=True)
```

```python
sns.kdeplot(x='area', data=df)
```
