merge two Dataframes      inner  left   right   outer merging 

```python
math =pd.DataFrame( {'std-num':[1,2,3,4,5], 'names':['ali', 'ahmad', 'iman', 'amir', 'sadra'] })
physics =pd.DataFrame( {'std-num':[1,2,3,4,5], 'names':['kamran', 'ahmad', 'hassan', 'amir', 'sadra'] })
math , physics
```

```python
pd.merge(math, physics, how='inner', on='names')           # only mergy those common names between dataframes
```

```python
pd.merge(math, physics, how='left', on='names')
```

```python
pd.merge(math, physics, how='right', on='names')
```

```python
pd.merge(math, physics, how='outer', on='names')
```
